```python
import requests, json
!pip install requests
# install requests (!pip install requests)
# goal: create a Pokemon... use functions 
# uising API data from the link provided
# Create a dict
# Pseudocode: 
# Get pkemon data: --send get request to url..
# parse response into JSON obj
# extract 'name', 'abilities', 'types', and 'weight' from the JSON obj
# return name,abilities,types, weights
# create pokemon dict:
# /initialize an empty pokemon_dict{}
# for names in pkmon names:
# call get_pkemon_data and get names...
# for type, if type not in pokemon_dict... add it.. initialize value as empty dict
# 
# create call function [run()]
```

```python
# # def get_pokemon_data(pokemon_name):
# response = requests.get('https://pokeapi.co/')
# print(response)

# # check if sucessful
# response.ok

def get_pokemon_data(pokemon_name):
    response = requests.get(f'https://pokeapi.co/api/v2/pokemon/{pokemon_name}')
    pokemon_data = response.json()
    
    name = pokemon_data['name']
    
    abilities = [ability['ability']['name'] for ability in pokemon_data['abilities']]
    
    types = [type_['type']['name'] for type_ in pokemon_data['types']]
    
    weight = pokemon_data['weight']
    return name, abilities, types, weight

#uses data from get_pokemon_data to create dict of pokemon charaterized by types
def create_pokemon_dict(pokemon_names): 
    pokemon_dict = {}
    
    for pokemon_name in pokemon_names:
        name, abilities, types, weight = get_pokemon_data(pokemon_name)
        
        for type_ in types:
            if type_ not in pokemon_dict:
                pokemon_dict[type_] = {}

            pokemon_dict[type_][name] = {
                'abilities': abilities,
                'weight': weight
            }
    return pokemon_dict

# fetch 20 pokemon_names from API--- pass names to create_pokemon_dict to create dictionary
response = requests.get('https://pokeapi.co/api/v2/pokemon?limit=20')
data = response.json()

pokemon_names = [pokemon['name'] for pokemon in data['results']] #accessing list of pokemon dict, looping over for each individual data
pokemon_dict = create_pokemon_dict(pokemon_names)
print(pokemon_dict)
```
