```python
# imports
import pandas as pd
#loading and reading data
nba_stats = pd.read_csv('../Pandas Project II/nbastats2018-2019.csv')
nba_stats.head(10)
```

```python
# print keys to get a fair idea of what my data contains
print(nba_stats.keys())
```

```python
# drop duplicates before starting? 
nba_stats_ready = nba_stats.drop_duplicates(subset = 'Name').reset_index(drop=True)
nba_stats_ready
```

```python
# using nba_stats_ready
# are there any missing values?finding missing values
nan_nba_stats = nba_stats_ready[nba_stats_ready.isnull().any(axis=1)]
nan_nba_stats
# replace all NaN with N/A for clarity sake
nba_stats = nan_nba_stats.fillna('N/A') #df returned to original name nba_stats
nba_stats['FG3%'] = nan_nba_stats['FG3%'].fillna(0)
# nba_stats['FG3%']
nba_stats
# READY TO BE ANALYZED
```

# What was the average age of player in the league?

```python
# using nba_stats
# age? group by age... find mean
# can find mean without groupby? 
#using groupby will result in getting age for each group instead of entire league?
#Pseudo: average_age = nba_stats..index age.mean()
mean_age = nba_stats['Age'].mean()
mean_age 
# result is float of more than two decimal places... round?
mean_age = mean_age.round()

print(f'The average age of player in the entire league is {mean_age}')
```

# What player scored the most points?

```python
# finding the top scorer in the league
# first, find the maximum number of points in league,
maximum_points = nba_stats['Points'].max()
# Use filter to get only the players with the most points
# top_scorers = check if data in nba_stats (points) == max points... put data here
top_scorers = nba_stats[nba_stats['Points'] == maximum_points]
# top_scorers
# get the name of the top scorer
top_scorer_name = top_scorers['Name'].tolist() #tolist will bring out the player name only..no index

top_scorer_point = top_scorers['Points'].tolist()
print(f'The player who scored the most points is {top_scorer_name} with {top_scorer_point} points')
```

# What player had the most blocks during the season? Was it a post player (F/C)?

```python
# what is the maximum number of blocks?
maximum_blocks = nba_stats['Blocks'].max()
# maximum_blocks
# Filter nba_stats to get only the player(s) with the most blocks
top_block_players = nba_stats[nba_stats['Blocks'] == maximum_blocks]
# top_block_players
#was it a top scorer?
top_block_player_names = top_block_players['Name']

# Check if the top blocker(s) is a post player (F/C)
#No column with player position data
# Cannot determine this without position data

print(f'The player who had the most blocks during the season is {top_block_player_names}')
print('The nba_stats does not include position data hence it cannot be determined whether the player was F/C')
```

# Based on the regular season, who had the best chance to win a title given their win percentage?

```python
# who as it teams? or players?
# if G is the total Games
# and WS is total wins
# then WS/G = win_percentage for each player
# create a new column called win_percentage ->
nba_stats['Win_percentage'] = nba_stats['WS'] / nba_stats['G']
# nba_stats

# Finding the maximum win percentage
max_win_percentage = nba_stats['Win_percentage'].max()

# Filter the DataFrame to include only the team(s)/row with the highest win percentage
best_chance = nba_stats[nba_stats['Win_percentage'] == max_win_percentage]
# best_teams
#Get the name of the team/player with the highest win percentage
best_chance_team = best_chance['Team'].tolist()
best_chance_player = best_chance['Name'].tolist()

print(f'From the data, we can deduce that {best_chance_player} from Team {best_chance_team} had the best chance to win a title')
```

```python
# if we are looking for the best team 
# then, maybe.. groupby teams
# aggregate [used to sum by team] ..calc total wins by teams
nba_teams_stats = nba_stats.groupby('Team')
nba_teams_stats = nba_teams_stats.agg({'WS': 'sum', 'G': 'sum'})
# nba_teams_stats 
# calc win percentage for each team
nba_teams_stats['team_win_percent'] = nba_teams_stats['WS'] / nba_teams_stats['G']

# find max win percent
maximum_win_percentage = nba_teams_stats['team_win_percent'].max()

# get team with  highest win percentage
best_team = nba_teams_stats[nba_teams_stats['team_win_percent'] == maximum_win_percentage]
# best_team
# getting team only                            
best_chance_team = best_team.index.tolist()
print(f'From the data, we can deduce that Team {best_chance_team} had the best chance to win a title')           
```

# What player had the best 3-pt percentage?

# What player given their player effiecency rating was the clutchest during the season?

```python
# cluchest player? using PER ...player efficiency ratings
# find max rating
# get row
# get name

highest_per = nba_stats['PER'].max()
cluchest_row = nba_stats[nba_stats['PER'] == highest_per]
cluchest_player  = cluchest_row['Name'].tolist()
cluchest_per = cluchest_row['PER'].tolist()
print(f'Cluchest player will be {cluchest_player} with PER at {cluchest_per}')
```

# What team had the youngest roster?

```python
#using team and ages..groupby
# find mean age per team arrange in ascending order
# index min... print

team_group = nba_stats.groupby('Team')['Age'].mean().sort_values(ascending = True)
team_mean_age = team_group.idxmin()

print(f'The {team_mean_age} had the youngest roster')
```

# At the end of a game, who WOULDN'T you want on the Free Throw Line?

```python
# where FT is Free Throw 
# find min ft..
# handle N/A data 
ft_check = nba_stats[nba_stats['FT%'] != 'N/A']
# ft_check['FT%']
minimum_ft = ft_check['FT%'].min()

minimum_ft_row = nba_stats[nba_stats['FT%'] == minimum_ft]

worst_ft_shooter = minimum_ft_row['Name'].tolist()
worst_ft = minimum_ft_row['FT%'].tolist()

print(f"At the end of a game, you wouldn't want {worst_ft_shooter} on the Free Throw Line.\n These players have a Free Throw percentage of {worst_ft} respectively")
```
